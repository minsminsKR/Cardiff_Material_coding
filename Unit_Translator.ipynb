```python
## Functions

# Check is number? For checking whether is number in front of unit.
def isNumber(s):
    try:
        float(s)
        return True
    except ValueError:
        return False

# Unit convert
def convert_length_real(number,unit,to_unit):
    D = {'km':10**6, 'm':10**3, 'cm':10, 'mm':1}
    output = float(number) * D[unit] / D[to_unit]
    print(output, to_unit)
#convert_length_real(1000,'m','cm')

def convert_time_real(number,unit,to_unit):
    D = {'h':3600, 'min':60, 'sec':1}
    output = float(number) * D[unit] / D[to_unit]
    print(output, to_unit)
#convert_time_real(2,'h','sec')

def convert_filesize_real(number,unit,to_unit):
    D = {'PB':1024**5, 'TB':1024**4, 'GB':1024**3, 'MB':1024**2, 'KB':1024, 'Byte':1}
    output = float(number) * D[unit] / D[to_unit]
    print(output, to_unit)
#convert_filesize_real(1024,'KB','TB')
```

```python
# Q2_example_text.txt

def unit_translator(filename, D):
    path = '/content/drive/MyDrive/' + filename
    f = open(path).read()
    print('Input text')
    print(f,type(f),'\n')
    text = []
    f_s = f.split(' ')
    for sentence in f_s:
        for word in sentence.split(' '):
            text.append(word)
        
    print('List of input text :')
    print(text,'\n')
    
    # Preprocessing specific words like 'mm.' which contains '.'
    including_point_length = list()
    including_point_filesize = list()
    including_point_time = list()
    
    for keys in D.keys():
        for a in Dic[keys].keys():
            if keys == 'length':
                including_point_length.append(a)
                including_point_length.append(a+'.')
            elif keys == 'filesize':
                including_point_filesize.append(a)
                including_point_filesize.append(a+'.')
            elif keys == 'time':
                including_point_time.append(a)
                including_point_time.append(a+'.')
    
    # exclude word itself
    def removing_point(index):
        try:
            if index == 'length':
                including_point_length.remove(D[index])
                including_point_length.remove(D[index]+'.')
            elif index == 'filesize':
                including_point_filesize.remove(D[index])
                including_point_filesize.remove(D[index]+'.')
            elif index == 'time':
                including_point_time.remove(D[index])
                including_point_time.remove(D[index]+'.')
        except KeyError:
            pass
        
    removing_point('length')
    removing_point('filesize')
    removing_point('time')

    for i, number in enumerate(text):
        if isNumber(number) == True:
            if text[i+1] in including_point_length:
                print('Before :',text[i],text[i+1])
                try:
                    text[i] = str(float(text[i])* Dic['length'][text[i+1]]/Dic['length'][D['length']])
                    text[i+1] = D['length']
                except KeyError:
                    text[i+1] = text[i+1][:-1]
                    text[i] = str(float(text[i])* Dic['length'][text[i+1]]/Dic['length'][D['length']])
                    text[i+1] = D['length'] + '.'
                print('After :',text[i],text[i+1],'\n')
            
            elif text[i+1] in including_point_filesize:
                print('Before :',text[i],text[i+1])
                try:
                    text[i] = str(float(text[i])* Dic['filesize'][text[i+1]]/Dic['filesize'][D['filesize']])
                    text[i+1] = D['filesize']
                except KeyError:
                    text[i+1] = text[i+1][:-1]
                    text[i] = str(float(text[i])* Dic['filesize'][text[i+1]]/Dic['filesize'][D['filesize']])
                    text[i+1] = D['filesize'] + '.'
                print('After :',text[i],text[i+1],'\n')
                
            elif text[i+1] in including_point_time:
                print('Before :',text[i],text[i+1])
                try:
                    text[i] = str(float(text[i])* Dic['time'][text[i+1]]/Dic['time'][D['time']])
                    text[i+1] = D['time']
                except KeyError:
                    text[i+1] = text[i+1][:-1]
                    text[i] = str(float(text[i])* Dic['time'][text[i+1]]/Dic['time'][D['time']])
                    text[i+1] = D['time'] + '.'
                print('After :',text[i],text[i+1],'\n')
                
    print('List of output :\n',text)   
    test_result = open(filename[:-4]+'_translated'+filename[-4:],'w')
    test_result.write(' '.join(text))
    test_result.close()
    
Dic = {'filesize':{'PB':1024**5, 'TB':1024**4, 'GB':1024**3, 'MB':1024**2, 'KB':1024, 'Byte':1},
          'length':{'km':10**6, 'm':10**3, 'cm':10, 'mm':1},
          'time':{'h':3600, 'min':60, 'sec':1}}
```

```python
unit_translator('Q2_example_text.txt',{'length':'m','filesize':'Byte','time':'h'})
```
