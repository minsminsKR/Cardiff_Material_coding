```python
def lambda_machine(filename):
    
    # Function which deletes specific value in the list
    def remove_values(the_list, val):
        return [value for value in the_list if value != val]
    
    # Load def.txt
    path = '/content/drive/My Drive/' + filename
    f = open(path).read()
    print('File name :', filename)
    print('Original form :\n',f,'\n')
    f_s = f.split(' ')
    new_list = list()

    # Delete 'def'
    for junk in f_s:
        if junk == 'def':
            f_s.remove('def')

    # Generate 'lambda'
    f_2 = list()
    for junk in f_s:
        for i, word in enumerate(junk):
            if word == '(':
                new_list.append(junk[:i])
                new_list.append('=')
                new_list.append('lambda')
                for i_2, word in enumerate(junk):
                    if word == ')':
                        new_list.append(junk[i+1:i_2])
                        new_list.append(':')
        f_2.append(junk)

    # Search formula after 'if'
    a=list()
    for i, word in enumerate(f_2):
        if word == 'if':
            for i_2, word_2 in enumerate(f_2):
                if word_2 == 'return':
                    for i_3, word_3 in enumerate(f_2):
                        if word_3 == 'else:\n':
                            a.append(f_2[i_2+1:i_3-3])
            
            # ['x', '+', 'y\n'] Delete the blank
            for x in a:
                if x == []:
                    a.remove(x)   
            b = list() 
            for x in a:
                for x_2 in x:
                    b.append(x_2)
            b = ''.join(b)[:-1]
            new_list.append(b)
            new_list.append('if')
    
            # If fomula
            if_list = list()
            if_list.append(f_2[i+1:])
            if_list2 = list()
            for x in if_list:
                for x_2 in x:
                    if_list2.append(x_2)
            if_fomula_raw = list()
            returnFlag = 0
            while returnFlag < 1:
                for i, word in enumerate(if_list2):
                    if word == 'return':
                        returnFlag += 1
                        if_fomula_raw.append(if_list2[:i])
                        break
            if_fomula = list()
            for x in if_fomula_raw:
                for x_2 in x:
                    if_fomula.append(x_2)
            if_fomula = remove_values(if_fomula,'')
            if_fomula = ' '.join(if_fomula)
            for i, word in enumerate(if_fomula):
                if word == ':':
                    if_fomula = if_fomula[:i]
            new_list.append(if_fomula)
        
            # Else_fomula       
            else_fomula = list()
            else_fomula_raw = list()
            for i, word in enumerate(f_2):
                if word == 'else:\n':
                    new_list.append('else')
                    else_fomula_raw.append(f_2[i+1:])
            for x in else_fomula_raw:
                for x2 in x:
                    else_fomula.append(x2)
            else_fomula = remove_values(else_fomula,'')
            for i in else_fomula:
                if i == 'return':
                    else_fomula.remove('return')
            else_fomula = ' '.join(else_fomula)
            new_list.append(else_fomula)

    # Simple converting formula when 'if' isn't there.
    if 'if' not in f_2:
        for i, word in enumerate(f_2):
            if word == 'return':
                new_list.append(' '.join(f_2[i+1:]))
    print('Changed title :',filename[:-4]+'_translate_to_lambda'+filename[-4:])
    print('Output :',' '.join(new_list))
    result = open(filename[:-4]+'_translate_to_lambda'+filename[-4:],'w')
    result.write(' '.join(new_list))
    result.close()
```

```python
lambda_machine('def_test1.txt')
```
